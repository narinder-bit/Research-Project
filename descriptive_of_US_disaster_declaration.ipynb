# Combined Datasets Analysis of Quarterly Percentages of Incident Types



```python

# Examine two datasets.
# -----------------------------------------------------------

# Put the datasets together.
combined_data = pd.concat([disaster_datacard, disaster5_datacard])

# By converting declaration_date to datetime format, the appropriate time zone can be set.
combined_data['declaration_date'] = pd.to_datetime(combined_data['declaration_date']).dt.tz_localize(None)

# Describe the many types of occurrences.
incident_types = combined_data['incident_type'].unique()

# Organize the quarterly percentages for each category of incidence in a dictionary.
quarterly_percentages = {}


#Determine Quarterly Percentages for Each Type of Incident
# -----------------------------------------------------------

# Determine quarterly percentages for every category of event.
for incident_type in incident_types:
    incident_type_data = combined_data[combined_data['incident_type'] == incident_type]
    quarter_totals = incident_type_data.groupby(incident_type_data['declaration_date'].dt.to_period('Q')).size()
    total_all = quarter_totals.sum()
    percentages = [quarter_total / total_all * 20 for quarter_total in quarter_totals]

    # Print out the incident type and the % length.
    print(f"Incident Type: {incident_type}, Length of Percentages: {len(percentages)}")

    quarterly_percentages[incident_type] = percentages


# Present the Results.
# -----------------------------------------------------------

# Determine quarterly percentages for every category of event.
for incident_type in incident_types:
    incident_type_data = combined_data[combined_data['incident_type'] == incident_type]
    quarter_totals = incident_type_data.groupby(incident_type_data['declaration_date'].dt.to_period('Q')).size()
    total_all = quarter_totals.sum()
    percentages = [quarter_total / total_all * 20 for quarter_total in quarter_totals]

    # Print out the incident type and the % length.
    print(f"Incident Type: {incident_type}, Length of Percentages: {len(percentages)}")

    quarterly_percentages[incident_type] = percentages


```

#Disaster Trends in States Over the Past Year: A Heatmap Analysis of Declarations


```python
# Construct a pivot table.
tp = combined_data.pivot_table(values='disaster_number', index='state', columns='fy_declared').fillna(0)

# Size the heatmap as desired.
plt.figure(figsize=(12, 8))

# Build the heatmap.
sns.heatmap(tp, cmap='coolwarm')

# Publish the heatmap as a picture.
plt.savefig('heatmap.png')

# Display the heatmap.
plt.show()

```

The frequency of disaster declarations across several states and fiscal years is visually represented by the heatmap. The heatmap's cells are color-coded according to the amount of disaster declarations, with each cell representing a particular state and fiscal year combination. Lighter colors denote lower declaration numbers, while darker colors denote higher declaration counts. You can instantly spot trends and patterns in each state's disaster declarations over time by looking at the heatmap. Understanding which states and fiscal years had more frequent or significant disaster declarations is made easier with the help of this visualization.
